```python
% matplotlib inline

from scikits.talkbox.features import mfcc
import numpy as np
import scipy
import matplotlib.pyplot as plt
import soundfile as sf
import pyaudio as pa
import os

random_state=24
```

```python
files = []
datapath = '../cry_data/devided_wav_data/'
# datapath = 'devided_wav_data2/'

for x in os.listdir(datapath):  
    if '.wav' in x:
        files.append(x)
```

```python
label_dict = {'hu':0, 'ti':1, 'dc':2}

X = []
y = []
for f in files:
    x, sample_rate = sf.read(datapath + f)
    if len(x) > 10000:
        X.append(x)
        if '-hu.' in f:
            y.append(label_dict['hu'])
        elif '-ti.' in f:
            y.append(label_dict['ti'])
        else:
            y.append(label_dict['dc'])
```

```python
f, t, Sxx = scipy.signal.spectrogram(X[0],fs=44100, nperseg=256)
```

```python
plt.figure()
plt.pcolormesh(t,f,Sxx,vmax=1e-6)
#plt.xlim([0,18])
plt.xlabel(u"time [sec]")
plt.ylabel(u"freqency [Hz]")
# plt.colorbar()
plt.show()
```
